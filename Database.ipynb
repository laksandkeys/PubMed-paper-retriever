```
def csv_to_table(filename):
  filepath = "/content/drive/My Drive/Fall20_IntrotoDS_Final_Project/" + filename
  df = pd.read_csv(filepath)
  conn = sqlite3.connect('/content/drive/My Drive/Fall20_IntrotoDS_Final_Project/project_part2.db')

  df.to_sql('pubmed_data',con=conn,if_exists='replace')
  conn.close()
  print("Done! CSV file is imported to a table")
```

```
def author_query(author_name):
  author_name = author_name.replace(" ","%")
  
  conn = sqlite3.connect('/content/drive/My Drive/Fall20_IntrotoDS_Final_Project/project_part2.db')
  c = conn.cursor()
  expression = "Select PMID, Title, AuthorList, PubDate From pubmed_data Where AuthorList LIKE '%" + str(author_name) + "%'"
  c.execute(expression)
  result = c.fetchall()
  conn.close()

  print(expression)
  return result
```
